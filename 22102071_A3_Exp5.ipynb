```python
# Generate a signal x1(t) = cos(6pit) and plot for 0<t<=1
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 1, .01)
x1 = np.cos(6 * np.pi * t)
plt.figure(figsize=(10, 5))
plt.plot(t, x1,label = 'cos(6pit)')
plt.xlabel('Time')
plt.ylabel('Amplitude)')
plt.title('Signal x1(t) = cos(6pit) 22102071')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot four periods of signal x2t = 3cos(3pi+pi/3)
t = np.arange(0, 4, .01)
x2 = 3*(np.cos(3*np.pi*t + np.pi/3))
plt.figure(figsize=(10, 5))
plt.plot(t, x2,label = 'x2(t)')
plt
plt.xlabel('Time')
plt.ylabel('Amplitude)')
plt.title('Signal x2(t) = 3cos(3pi+pi/3) 22102071')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the exponential signals x3t = 3e^0.4t and x4t = 2e^-0.9t
np.arange(0, 1, .01)
x3 = 3*np.exp(0.4*t)
x4 = 2*np.exp(-0.9*t)
plt.figure(figsize=(10, 5))
plt.plot(t, x3,label = 'x3(t)')
plt.plot(t, x4,label = 'x4(t)')
plt.xlabel('Time')
plt.ylabel('Amplitude)')
plt.title('Signal x3(t) = 3e^0.4t and x4t = 2e^-0.9t 22102071')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Write and generate the follwing sequence
# gn = 5, n<=0 ; gn = 5-3n, 0<n<=4; gn = -23+n^2, 4<n<=8; gn = 41,n>8
n = np.arange(-12, 13)
g = 5*(n<=0) + (5-(3*n))*((4>=n)*(n>8)) + (-23+pow(n,2))*((8>=n)*(n>4)) + 41*(n>8)
plt.figure(figsize=(10, 5))
plt.stem(n, g,label = 'g(n)')
plt.xlabel('n')
plt.ylabel('g(n)')
plt.title('Sequence g(n) 22102071')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate s signal x5t by adding x1t and x2t
t = np.arange(0, 3, .01)
x1 = np.cos(6 * np.pi * t)
x2 = 3*(np.cos(3*np.pi*t + np.pi/3))
x5  = x1 + x2
plt.figure(figsize=(10, 5))
plt.plot(t, x5,label = 'x5(t)')
plt
plt.xlabel('Time')
plt.ylabel('Amplitude)')
plt.title('Signal x5(t) = x1(t) + x2(t) 22102071')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Find even odd component x3 t
t = np.arange(-2, 2, .2)
def x3(t):
  return 3*np.exp(0.4*t)
plt.plot(t,x3(t),label = 'x3(t)')
plt.xlabel('Time')
plt.ylabel('Even and Odd Compnents')
x3_even = 0.5*(x3(t) + x3(-t))
x3_odd = 0.5*(x3(t) - x3(-t))
plt.plot(t,x3_even,"r",label = 'x3_even(t)')
plt.plot(t,x3_odd,"g",label = 'x3_odd(t)')
plt.title('Signal x3(t) = 3e^0.4t 22102071')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# generate xt = te^-t,-1<t<=1
# x(t);x(-t);x(t/5);x(t-3);x(1-2t);x(5t)
t = np.arange(-1, 1, .01)
x = t*np.exp(-t)
x1 = -t*np.exp(-t)
x2 = t/5*np.exp(-t)
x3 = t-3*np.exp(-t)
x4 = 1-2*t*np.exp(-t)
x5 = 5*t*np.exp(-t)
plt.figure(figsize=(10, 5))
plt.plot(t,x,label="x(t)")
plt.plot(t,x1,label="x(-t)")
plt.plot(t,x2,label="x(t/5)")
plt.plot(t,x3,label="x(t-3)")
plt.plot(t,x4,label="x(1-2t)")
plt.plot(t,x5,label="x(5t)")
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.title('Signal xt = te^-t,-1<t<=1 22102071')
plt.legend()
plt.grid(True)
plt.show()
```
