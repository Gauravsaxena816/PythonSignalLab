### Experiment 3
**Aim** : DT Convolution To calculate the convolution sum of two discrete time signals.

**Theory** :Convolution is a mathematical operation used to express the relation between input and output of an LTI system. It relates input, output and impulse response of an LTI system. The discrete time
convolution of two sequences can be defined as x[n] ∗ h[n]= Σ∞
x[k]h[n − k]

```python
# experiment 3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# to plot a convolution of two discrete signal
# 22102071
# Problem 1: Compute the convolution of the following discrete time sequences
#x1=([1.0, 2.0, 3.0]) x2=([0.0, 1.0, 0.5,1.0])
x1 = ([1,3,4,5,6,7])
x2 = ([1,2,3,4,5,6])
y1 = np.convolve(x1,x2)
plt.subplot(221)
plt.stem(y1,use_line_collection=True)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('DT Convolution of x1 and x2 - 22102071')
# Problem 2 : compute the convolution of the discrete time sequence
x21 =([-1.0,2.0,-3.0])
x22 = ([0.0,-1.0,0.5,1.0])
y2 = np.convolve(x21,x22)
l21 = np.size(x21)
l22 = np.size(x22)
no = np.arange(0,l21+l22-1,1)
plt.subplot(222)
plt.stem(no,y2,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('y')
plt.title('DT Convolution of x1 and x2 - 22102071')
# Problem 3 compute the convolution of the discrete time sequence
# x1[n1]=([-1.0, 2.0, -3.0]), n1 =-1 to 1 x2[n2]=([0.0, -1.0, 0.5,1.0]), n2 = ([-2, -1, 0, 1])
x31 =([-1.0,2.0,-3.0])
x32 = ([0.0,-1.0,0.5,1.0])
n31 = ([-1,0,1])
n32 = ([-2,-1,0,1])
y3 = np.convolve(x31,x32)
l31 = np.size(x31)
l32 = np.size(x32)
n3 = np.arange(0,l32+l31-1,1)
plt.subplot(223)
plt.stem(n3,y3,use_line_collection=True)
plt.xlabel('n')
plt.ylabel('y')
plt.title('DT Convolution of x1 and x2 - 22102071')
# Problem 4: Compute the convolution of the following discrete time sequences x1[n]=[0,1,0,1,0,1] x2[n]=[1,0,0,0]
x41 = ([0,1,0,1,0,1])
x42 = ([1,0,0,0])
y4 = np.convolve(x41,x42)
l41 = np.size(x41)
l42 = np.size(x42)
n4 = np.arange(0,l41+l42-1,1)
plt.subplot(224)
plt.stem(n4,y4)
plt.xlabel('n')
plt.ylabel('y')
plt.title('DT Convolution of x1 and x2 - 22102071')
plt.figure(figsize=(500,50))
plt.tight_layout(pad=10, w_pad=1, h_pad=10)
```

```python
# POST LAB
# Problem 1: Compute the convolution of x1[n]=u[n]-u[n-3] and x2[n]=u[n]-u[n-2].
n = np.arange(0,5,1)
def u(n):
    return np.where(n >= 0, 1, 0)
x1 = u(n) - u(n-3)
x2 = u(n) - u(n-2)
y = np.convolve(x1,x2)
plt.stem(y)
plt.xlabel('n')
plt.ylabel('y')
plt.title('DT Convolution of x1 and x2 - 22102071')
```
