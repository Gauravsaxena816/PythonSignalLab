```python
# to generate a discrete time sine signal
dt=0.11
t=np.arange(0,5,dt)
s=np.sin(2*np.pi*t)
plt.stem(t,s)
plt.title("discrete time sine signal")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
# to generate a discrete time unit-impulse and unit-step function
n= np.linspace(-5,5,11)
delta = 1*(n==0)
u = 1*(n>=0)
plt.subplot(2,1,1)
plt.stem(n,delta,label="delta(n)",use_line_collection=True)
plt.xlabel("n")
plt.ylabel("delta(n)")
plt.legend()
plt.grid()
plt.subplot(2,1,2)
plt.stem(n,u,label="u(n)",use_line_collection=True)
plt.xlabel("n")
plt.ylabel("u(n)")
plt.legend()
plt.grid()
plt.show()
```

```python
# To create two subplots
#creating function for generating signal using diffrent time ranges
def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure()
#subplot matrix size 2 x 1, window 1
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.grid()
#subplot matrix size 2 x 1, window 2
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.grid()
plt.show()
```

```python
# to plot sine wave
x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)
plt.subplot(3,1,1)
plt.plot(x, y,'blue', label='sine')
plt.legend()
plt.grid()
plt.show()
# to plot negative exponential
x = np.arange(0, 5, 0.1)
y = np.exp(-x)
plt.subplot(312)
plt.plot(x, y,'red',label='negative exponential')
plt.legend()
plt.grid()
plt.show()
# to plot damped sine wave
x = np.arange(0, 5, 0.1)
y = np.sin(2 * np.pi * x) * np.exp(-x)
plt.subplot(313)
plt.plot(x, y,'green',label='damped sine wave')
plt.legend()
plt.grid()
plt.show()
```

```python
# Write a program to plot two sequences given below in one figure window. y[n] = u[n+3] - u[n-4], u[n] is unit step sequence y[n] = r[n], n = -5:20, r[n] is unit ramp sequence
n = np.arange(-5, 21)


def u(n):
    return np.where(n >= 0, 1, 0)
def r(n):
    return np.where(n >= 0, n, 0)


y = u(n + 3) - u(n - 4)
r_n = r(n)

# Plotting the sequences
plt.figure(figsize=(12, 6))


plt.subplot(2, 1, 1)
plt.stem(n, y)
plt.title('Sequence y[n] = u[n+3] - u[n-4]')
plt.xlabel('n')
plt.ylabel('y[n]')


plt.subplot(2, 1, 2)
plt.stem(n, r_n)
plt.title('Sequence r[n]')
plt.xlabel('n')
plt.ylabel('r[n]')

plt.tight_layout()
plt.show()
```
