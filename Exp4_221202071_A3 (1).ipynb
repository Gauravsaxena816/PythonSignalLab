### Aim : Writing codes to compute DFT (Discrete Fourier Transform) and IDFT (Inverse Discrete Fourier Transform) for the spectral analysis of signals.




```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.fft
```

```python
# Computation of DFT
# Problem 1: Compute DFT of the sequence x1 = [0, 4, 2, 0]
x1=([0,4,2,0])
#DFT computation of x1
dft=scipy.fft.fft(x1)
plt.subplot(2, 1, 1)
plt.stem(dft.real, use_line_collection = True)
plt.xlabel('k')
# naming the y axis
plt.ylabel('Real{x[k]}')
# Title to plot
plt.title('Real part of DFT 22102071')
plt.show()


plt.subplot(2, 1, 2)
plt.stem(dft.imag, use_line_collection = True)
plt.xlabel('k')
plt.ylabel('Img{X{k}}')
# Title to plot
plt.title('Imaginary Part of DFT 221202071')
plt.show()
print('DFT X[k] =',dft)
```

```python
# Problem 2: Computation of IDFT of X[k] = [6, 2  4j, 2, 2 + 4j]
Xk=[6,-2-4j,-2,-2+4j]
#IDFTcomputation of Xk
idft=scipy.fft.ifft(Xk)
plt.subplot(2, 1, 1)
plt.stem(idft.real, use_line_collection = True) # naming the x axis
plt.xlabel('n')
# naming the y axis
plt.ylabel('Real{x[n]}')
plt.title('Real part of IDFT 22102071')
plt.show()
plt.subplot(2, 1, 2)
plt.stem(idft.imag, use_line_collection = True)
plt.xlabel('n')
# naming the y axis
plt.ylabel('Img{x[n]}')
plt.title('Imaginary Part of IDFT 22102071')
plt.show()
print('IDFT x[n] =',idft)
```

```python
# Problem 3: Compute the DFT of the sin(t)
from scipy.fft import fft, fftfreq, fftshift
N = 32
t = np.arange(N)
x=np.sin(t)
sp = fftshift(fft(x))
freq = fftshift(fftfreq(N))
plt.figure(figsize=(4,8)) #Ploting x(t)
plt.subplot(211)
plt.grid()
plt.plot(t, x)
plt.subplot(212)
plt.grid()
plt.plot(freq, abs(sp)) # Real Part
# naming the x axis
plt.xlabel('k')
# naming the y axis
plt.ylabel('Magnitude Spectrum {|X[k]|} 22102071')
plt.show()
```

### Unsolved Problem
Problem 1: Compute and plot DFT of x[n] = u[n]  u[n  3]

```python
# DFT of x[n] = u[n]  u[n  3]
N = 10
n = np.arange(N)
x = np.heaviside(n, 1) - np.heaviside(n-3, 1)
X = scipy.fft.fft(x)
plt.stem(np.abs(X))
plt.xlabel('k')
plt.ylabel('|X[k]|')
plt.title('Magnitude of DFT 22102071')
plt.show()
```

```python
# Problem 2: Compute and plot DFT of x[n] = 2u[n]  4u[n  2] + 2u[n  6]
N = 10
n = np.arange(N)
x = 2*np.heaviside(n, 1) - 4*np.heaviside(n-2, 1) + 2*np.heaviside(n-6, 1)
X = scipy.fft.fft(x)
plt.stem(np.abs(X))
plt.xlabel('k')
plt.ylabel('|X[k]|')
plt.title('Magnitude of DFT 22102071')
plt.show()
```
